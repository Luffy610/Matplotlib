```python
data = pd.read_csv('07-data.csv')
```

```python
data.head()
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, edgecolors='black', linewidths=1, alpha=0.75)
```

```python
plt.scatter(view_count, likes, edgecolors='black', linewidths=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
plt.scatter(view_count, likes, c=ratio,cmap='summer', edgecolors='black', linewidths=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')
cbar = plt.colorbar()
cbar.set_label("Likes Dislikes Color Ratio")
plt.show()
```
