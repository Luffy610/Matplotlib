```python
import pandas as pd
```

```python
df = pd.read_csv("02-data.csv")
```

```python
df.set_index('Responder_id', inplace=True)
```

```python
df.head()
```

```python
df_languages = df['LanguagesWorkedWith'].str.split(";")
```

```python
df_languages.head()
```

```python
from collections import Counter
```

```python
lang_counter = Counter()
```

```python
for responses in df_languages:
    lang_counter.update(responses)
```

```python
lang_counter
```

```python
most_common = lang_counter.most_common(15)
```

```python
most_common
```

```python
lang = []
count = []
for item in most_common:
    lang.append(item[0])
    count.append(item[1])
```

```python
lang
```

```python
count
```

```python
lang.reverse()
count.reverse()
plt.barh(lang, count)
plt.title("Most Popular Languages")
plt.xlabel("No. of People used")
```
